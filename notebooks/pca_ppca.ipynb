# Probabilistic PCA

## Imports

```python
from ppca import PPCA
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from ppca import PCA
```

## Datasets

```python
# Loading data

iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

## PCA vs PPCA vs PPCA generated comparison

```python
# PPCA parameters
n_components = 2
method = 'em'
epochs = 100
```

```python
# PPCA
ppca = PPCA(n_components=n_components, method=method, max_iter=epochs)
Xt = ppca.fit_transform(X)

# PCA for comparison
pca = PCA(2)
Xt_pca = pca.fit_transform(X)
```

```python
# Grouped plots in one figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Left: PCA projection
scatter1 = axs[0].scatter(Xt_pca[:, 0], Xt_pca[:, 1], c=y, cmap='viridis', alpha=0.8)
axs[0].set_title(f'PCA Projection (2D) {n_components} components')
axs[0].set_xlabel('Principal Component 1')
axs[0].set_ylabel('Principal Component 2')

# Middle: original PPCA projection
scatter2 = axs[1].scatter(Xt[:, 0], Xt[:, 1], c=y, cmap='viridis', alpha=0.8)
axs[1].set_title(f'PPCA Projection (2D) {n_components} components')
axs[1].set_xlabel('Principal Component 1')
axs[1].set_ylabel('Principal Component 2')

plt.tight_layout()
plt.show()
```
